```python
# Get hist0rical data
hist_spx = api.priceHistory.bySymbol("SPX", 
                                     periodType='month', 
                                     period=6, 
                                     frequencyType='daily', 
                                     frequency=1, 
                                     startDate=None,
                                     endDate=None, 
                                     needExtendedHoursData=False, 
                                     needPreviousClose=True).json()
```

```python
spx_df = pd.DataFrame(hist_spx['candles'])
spx_df['datetime'] = pd.to_datetime(spx_df['datetime'],unit='ms')
spx_df.set_index('datetime')
```
